```python
import pylab
import pandas as pd
```

```python
log_df = pd.read_csv("/home/datascience/Downloads/wc_day6_1_sample.csv",
                     names=['ClientID', 'Date', 'Time', 'URL', 'ResponseCode', 'Size'],
                     na_values=['-'])
```

```python
#Question 2:
%matplotlib inline
# show the percentage of each response code
import matplotlib.pyplot as plt
```

```python
q2_df = log_df[log_df['Date'] == '01/May/1998']
q2_df['DateTime'] = pd.to_datetime(q2_df.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
hour_grouped = q2_df.groupby(lambda x: q2_df['DateTime'][x].hour)
```

```python
hour_grouped['ClientID'].nunique().plot(kind='bar')
```

```python
#Question 3:

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = hour_grouped.size().index

ax1.plot(x, hour_grouped['ClientID'].nunique(), 'g-')
ax2.plot(x, hour_grouped['Size'].sum(), 'r-')

ax1.set_xlabel('Hour of the day')
ax1.set_ylabel('# of Distinct Users', color='g')
ax2.set_ylabel('Size of traffic handled', color='r')
```

```python
#Question 4:
client=sorted(log_df['ClientID'].unique())[100:200]
df=log_df[log_df['ClientID'].isin(client)]
df['DateTime'] = pd.to_datetime(df.apply(lambda row: row['Date'] + ' ' + row['Time'], axis=1))
df['hour'] = df.DateTime.apply(lambda x: x.hour)
df.plot(kind='scatter', x='hour', y='ClientID')
```
